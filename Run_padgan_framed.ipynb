```python
import Padgan_variants
import load_data
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

```python
x_scaled, y, _, xscaler = load_data.load_framed_dataset("r", onehot = True, scaled = True, augmented = True)

yscaler = StandardScaler()
yscaler.fit(y)
y_scaled = yscaler.transform(y)
# y_scaled=pd.DataFrame(y_scaled, columns=y.columns, index=y.index)
x_scaled=x_scaled.values
```

```python
regressor=tf.keras.models.load_model("./Trained Models/BestNN")

def tf_inv_scale(y_scaled):
    scale=tf.cast(tf.convert_to_tensor(y_scaler.scale_), "float32")
    mean=tf.cast(tf.convert_to_tensor(y_scaler.mean_), "float32")
    return tf.cast(y_scaled*scale+mean, "float32")

def regressor_scaled(y):
    return tf_inv_scale(regressor(y))

classifier=tf.keras.models.load_model("./Trained Models/BestNN_Classifier_final")

config_params = [True, True, True, classifier, regressor]
train_params = [5, 0.5, 4, 5000]
DTAI_params= ["auto", "auto", "auto"]
trained_model = Padgan_variants.train_padgan(x_scaled, None, y_scaled, config_params, train_params, DTAI_params, None, None)
```

```python
#Print the first 5 datapoints
print(xscaler.inverse_transform(x_scaled[:5, :]))
#Generate 5 samples
print(xscaler.inverse_transform(trained_model.generate(5)))
```
